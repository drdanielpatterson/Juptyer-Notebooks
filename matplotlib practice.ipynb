```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(11, 7), linewidth = 3, edgecolor = 'blue')

fig.set_edgecolor('red')
fig.set_facecolor('gray')
#ax.set_facecolor('blue')

grades = [3, 4, 5, 6]
pa = [40, 45, 30, 55]

bar_fill_color = '#219ebc'
bar_edge_color = '#fb8500'
bar_width = 0.5

ax.bar(grades, pa, color = bar_fill_color,  edgecolor = bar_edge_color, width = bar_width)

j = 0
for i in grades:
    plt.text(i,pa[j]+2,pa[j], ha = 'center')
    j = j + 1

ax.set_xlabel('A Measurement', fontname="Matura MT Script Capitals", fontsize=24)
ax.set_ylabel('A Measurement')
ax.set_title('A Graph')

ax.set_xticks([3,4,5,6])
ax.set_yticks([20, 40, 60, 80])
ax.set_ylim(0, 100)

#ax.margins(x = 0.5, y = 0.5)
#fig.tight_layout(pad=0, h_pad=0, w_pad=0, rect = (0,0,0,1))
#plt.tight_layout(rect = (0,0,0,0.2))
fig.tight_layout()
plt.show()
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
